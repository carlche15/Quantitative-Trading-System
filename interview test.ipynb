```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def coupon_bond(yd, coupon, term):
    N = 100.
    cf = np.ones(term)*coupon*N
    cf[-1]+=N
    pv = np.sum(cf/((1+yd)**(np.arange(1,term+1))))
    return pv
    
```

```python
yd_range = np.linspace(0.02,0.2,50)
plt.plot(yd_range, [coupon_bond(y,0.1,10) for y in yd_range])
```
